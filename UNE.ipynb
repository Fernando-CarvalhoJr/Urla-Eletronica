```python
candidatos = {'4002': ('Carol', 'CIA'), '8922': ('Jamelão', 'BOM'), '9999': ('Fernando', 'DIA')}

def inic_voto (candidatos):
    candidatos_Carol = 0
    candidatos_Jamelão = 0
    candidatos_Fernando = 0
    
    
    for numero, info_cand in candidatos.items():
        nome, partido = info_cand
        
        while True:
            print("Opções do eleitor\n")
            print("1. Selecionar Voto")
            print("2. Voto Branco")
            print("3. Sair")

            escolha = input('Selecione uma opção: ')


            if escolha == '1':

                voto = input('Selecione o número do seu candidato: ')

                if voto == '4002':
                    candidatos_Carol += 1
                    print('\nVocê votou na Carol do partido CIA\n')

                elif voto == '8922':
                    candidatos_Jamelão += 1
                    print('\nVocê votou na Jamelão do partido BOM\n')
                    
                elif voto == '9999':
                    candidatos_Fernando += 1
                    print('\nVocê votou na Fernando do partido DIA\n')
                    
                else:
                    print('\nCanditato não encontrado, tente novamente.\n')

            elif escolha == '2':

                print('\nSeu voto foi branco.\n')

            elif escolha == '3':
                print('\nEncerrando e calculando votos...\n')
                break
        break
        
    total_votos = candidatos_Carol + candidatos_Jamelão + candidatos_Fernando
    
    if candidatos_Carol > 0:
        votos_Carol = (candidatos_Carol / total_votos) * 100
        print(f'Carol recebeu {candidatos_Carol} votos tento a porcentagem de {votos_Carol}%')
    else:
        print('Não houve votos para Carol')
        
    if candidatos_Jamelão > 0:
        votos_Jamelão = (candidatos_Jamelão / total_votos) * 100
        print(f'Jamelão recebeu {candidatos_Jamelão} votos tento a porcentagem de: {votos_Jamelão}%')
    else:
        print('Não houve votos para Jamelão')
    
    if candidatos_Fernando > 0:
        votos_Fernando = (candidatos_Fernando / total_votos) * 100
        print(f'Fernando recebeu {candidatos_Fernando} votos tento a porcentagem de: {votos_Fernando}%')
    else:
        print('Não houve votos para Fernando')

inic_voto (candidatos)


    
     
```
